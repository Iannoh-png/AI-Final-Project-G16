```python
# Convert Quantity to numeric first (ignoring text errors)
data['Quantity_numeric'] = pd.to_numeric(data['Quantity'], errors='coerce')

# Find Negative quantities
print(data[data['Quantity_numeric'] < 0])
```

```python
# Try to convert dates and find ones that fail
data['OrderDate_converted'] = pd.to_datetime(data['OrderDate'], errors='coerce')

# Show rows where date conversion failed
print(data[data['OrderDate_converted'].isna()])
```
